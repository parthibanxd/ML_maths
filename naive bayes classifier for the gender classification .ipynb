```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Load dataset
file_path = r"C:\Users\Parthiban\Documents\Model lab\gender_classification_v7.csv"
df = pd.read_csv(file_path)

# Display first few rows
print("Dataset Preview:\n", df.head())

# Step 2: Split features (X) and label (y)
X = df.drop('gender', axis=1)
y = df['gender']

# Step 3: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Step 4: Train Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# Step 5: Predict on test data
y_pred = model.predict(X_test)

# Step 6: Evaluate performance
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# ✅ Step 7: Predict for new profiles (example input)
# Make sure these columns match your dataset
new_profiles = pd.DataFrame({
    'long_hair': [1, 0],
    'forehead_width_cm': [14, 12],
    'forehead_height_cm': [5.5, 4.9],
    'nose_wide': [0, 1],
    'nose_long': [0, 1],
    'lips_thin': [1, 0],
    'distance_nose_to_lip_long': [0, 1]
})

predictions = model.predict(new_profiles)
print("\nPredicted Genders for New Profiles:", predictions)
```
